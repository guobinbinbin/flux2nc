```python
import os
import shutil
from pathlib import Path
from flux2nc import flux2nc
```

```python
model_path = Path(r'D:\VMWare_Share\centos_share\VIC4_RUN\kge\run\GA')

output_path = Path(r'D:\PROJECTS\Xiangjiang_VIC\vic_fluxes_nc_output\GA')

for m in model_path.glob('*/'):
    best_vic_output = next((d for d in m.joinpath('output_best').iterdir() if d.is_dir()), None).joinpath('vic_output')

    # print(m.__str__() + '/')
    # print(output_path.__str__() + '/')
    #
    # break


    # create output path
    out_model_path = output_path.joinpath(m.name)
    if os.path.exists(out_model_path):
        shutil.rmtree(out_model_path)

    os.mkdir(out_model_path)

    for soil_lyr in [1, 2, 3]:

        print(f'{m.name} - soil_lyr{soil_lyr} Processing...')

        flux2nc(
            influxes= best_vic_output.__str__() + '/',
            outpath= out_model_path.__str__() + '/',
            var=6,
            camada=soil_lyr,
            start_year=2001,
            end_year=2020
        )


    for p in [1, 2, 3, 4, 5]:

        print(f'{m.name} - {p} Processing...')

        flux2nc(
            influxes= best_vic_output.__str__() + '/',
            outpath= out_model_path.__str__() + '/',
            var=p,
            camada=1,
            start_year=2001,
            end_year=2020
        )





    #
    #
    #
    # print(out_model_path)

    # break


    # print(best_vic_output)

```
